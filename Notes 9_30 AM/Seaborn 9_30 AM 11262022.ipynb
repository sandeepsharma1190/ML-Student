```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.lineplot(x = 'timepoint', y = 'signal', data = fmri)
```

```python
sns.barplot(x = 'timepoint', y = 'signal', data = fmri)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['tip'])
```

```python
sns.distplot(tips['tip'], kde = False)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, color = 'r')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, color = 'r', kind = 'kde')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, color = 'r', kind = 'hist')
```

```python
sns.pairplot(data = tips, hue = 'day')
```
