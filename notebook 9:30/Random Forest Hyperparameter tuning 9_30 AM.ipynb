```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv(r'C:\Users\LENOVO\Downloads\Datasets\insurance - Regression.csv')
```

```python
df = dataset.copy()
df.head()
```

```python
# df.isna().sum()
```

```python
df1 = pd.get_dummies(df, columns = ['sex', 'smoker', 'region'], drop_first = True)
df1.head()
```

```python
df1.columns
```

```python
X = df1.drop('charges', axis = 1)
y = df1.drop(['age', 'bmi', 'children', 'sex_male', 'smoker_yes',
              'region_northwest', 'region_southeast', 'region_southwest'], axis = 1)
```

```python
X.head()
```

```python
X.children.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X[['age', 'bmi']] = sc.fit_transform(X[['age', 'bmi']])
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = .2, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(random_state = 0)
rfr.fit(X_train, y_train)
```

```python
y_pred = rfr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2_score(y_test, y_pred)
```

```python
importance = rfr.feature_importances_
feat_name = X.columns
tree_result = pd.DataFrame({'Feature':feat_name, 'importance':importance})
tree_result_sort = tree_result.sort_values(by = 'importance', ascending = False)
```

```python
tree_result_sort
```

## Random Hyperparameter Grid

```python
from sklearn.model_selection import RandomizedSearchCV
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
max_depth = [int(x) for x in np.linspace(start = 10, stop = 100, num = 10)]
min_samples_split = [2,5,10]
min_samples_leaf = [1,2,4]
max_features= ['sqrt', 'log2']
bootstrap = [True, False]

random_grid = {'n_estimators' : n_estimators,
              'max_depth':max_depth,
              'min_samples_split':min_samples_split,
              'min_samples_leaf': min_samples_leaf,
              'max_features':max_features,
              'bootstrap':bootstrap}
```

```python
random_grid
```

```python
%%time
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(random_state = 0)
rfr_random = RandomizedSearchCV(estimator = rfr, param_distributions = random_grid,
                               n_iter = 5, cv = 3, verbose = 3)
rfr_random.fit(X_train, y_train)
```

```python
rfr_random.best_params_
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators = 1000, min_samples_split = 5, 
                            min_samples_leaf = 4, max_features = 'log2',
                            max_depth = 50, bootstrap = False)
rfr.fit(X_train, y_train)
```

```python
y_pred = rfr.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

```python
importance = rfr.feature_importances_
feat_name = X.columns
tree_result = pd.DataFrame({'Feature':feat_name, 'importance':importance})
tree_result_sort = tree_result.sort_values(by = 'importance', ascending = False)
```

```python
tree_result_sort
```
